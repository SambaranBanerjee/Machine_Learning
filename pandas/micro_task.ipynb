```python
import pandas as pd
import seaborn as sns
pd.set_option("display.max_rows", 5)
titles = pd.read_csv("../datasets/netflix_titles.csv", index_col=0)

#from learntools.core import binder; binder.bind(globals())
#from learntools.pandas.summary_functions_and_maps import *
print("Setup complete.")

titles.head()
```

```python
#Finding most common director
most_common_director = titles[['director']].value_counts().head(1)
most_common_director
```

```python
titles['date_added'] = titles['date_added'].str.strip() 
```

```python
#the year with the most content added
titles['date_added'] = pd.to_datetime(titles['date_added'])
year_column = titles['date_added'].dt.year
print(year_column.value_counts().idxmax())
```

```python
#the ratio of Movies to TV Shows
no_of_movies = titles['type'].map(lambda p: p == 'Movie')
no_of_TV_shows = titles['type'].map(lambda p: p == 'TV Show')
print((no_of_movies.sum())/(no_of_TV_shows.sum()))
```

```python
#the ratio of Movies to TV shows
#value_counts: this creates a single series where indexes are the unique categories and their count. in this case Movie and TV Show

type_of_content = titles['type'].value_counts(normalize=True)
print(type_of_content['Movie'])
print(type_of_content['TV Show'])
print(type_of_content['Movie']/type_of_content['TV Show'])
```
