```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_1 = np.linspace(0, 5, 10)#linespace[start, end, num] Returns an array of numbers between start and end with num being the number of outputs to be shown (default 50)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title("Distribution of X and Y")
plt.xlabel("X values")
plt.ylabel("Y values") 
```

```python
#Sub Plot
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'g')
```

```python
#Figure
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9]) #add_axes(left, bottom, width, height)
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title("Days Squared Chart")
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1, x_1, label='x^2/x')
axes_1.legend(loc=0)
```
