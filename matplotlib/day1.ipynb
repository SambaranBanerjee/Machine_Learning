```python
import pandas as pd
pd.set_option("display.max_rows", 5)
titles = pd.read_csv("../datasets/netflix_titles.csv", index_col=0)

print("Setup complete.")

titles.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Whenever an expert creates a plot for a report or portfolio, 
#they never stop at just the title. They always check the "Big Three":

#Labels: Are the Axes clearly named?
#Spines: Does the chat look boxy?(Use sns.despine() to remove the top and right borders for a modern, clean look)
#Context: Is the font readable size?(Use sns.set_theme(style='whitegrid') at the start of your script)
```

```python
sns.set_theme(style="white")
sns.countplot(data=titles, x='type', hue='type', palette='viridis', legend=False)
sns.despine()
plt.title("Distribution of Movies vs TV Shows", fontsize = 14, pad=20)
plt.xlabel("Content Type")
plt.ylabel("Total Count")
```
